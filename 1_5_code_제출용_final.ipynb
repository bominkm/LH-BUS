# 3. 함수 정리

```python
#Pydeck 사용을 위한 함수 정의
import geopandas as gpd 
import shapely # Shapely 형태의 데이터를 받아 내부 좌표들을 List안에 반환합니다. 

def line_string_to_coordinates(line_string): 
    if isinstance(line_string, shapely.geometry.linestring.LineString): 
        lon, lat = line_string.xy 
        return [[x, y] for x, y in zip(lon, lat)] 
    elif isinstance(line_string, shapely.geometry.multilinestring.MultiLineString): 
        ret = [] 
        for i in range(len(line_string)): 
            lon, lat = line_string[i].xy 
            for x, y in zip(lon, lat): 
                ret.append([x, y])
        return ret 

def multipolygon_to_coordinates(x): 
    lon, lat = x[0].exterior.xy 
    return [[x, y] for x, y in zip(lon, lat)] 

def polygon_to_coordinates(x): 
    lon, lat = x.exterior.xy 
    return [[x, y] for x, y in zip(lon, lat)] 
```

```python
# 중심점 찾는 함수
def multipolygon_to_center_coordinates(x): 
    lon, lat = x[0].centroid.xy 
    return [[x, y] for x, y in zip(lon, lat)]
def polygon_to_center_coordinates(x): 
    lon, lat = x.centroid.xy
    return [[x, y] for x, y in zip(lon, lat)] 
```

```python
# 100X100 grid에서 central point 찾기   
def multipolygon_to_center(x):
    list1 = []
    list2 = []
    x['coord_cent']=0
    x['geo_cent']=0 
    for i in x['geometry']:
        cent = [[i[0].centroid.coords[0][0],i[0].centroid.coords[0][1]]]
        list1.append(cent)
        list2.append(Point(cent[0]))
    x['coord_cent'] = pd.DataFrame(list1)
    x['geo_cent'] = pd.DataFrame(list2)
    return x

def polygon_to_center(x):
    list1 = []
    list2 = []
    x['coord_cent']=0
    x['geo_cent']=0 
    for i in x['geometry']:
        cent = [[i.centroid.coords[0][0],i.centroid.coords[0][1]]]
        list1.append(cent)
        list2.append(Point(cent[0]))
    x['coord_cent'] = pd.DataFrame(list1)
    x['geo_cent'] = pd.DataFrame(list2)
    return x
```

```python
#link_id를 df_23 link_id에 맞춰주는 함수
def link_id_change(data):
    data = pd.DataFrame(data).reset_index()
    del data['index']
    for i in range(len(data)):
        data['link_id'][i] = float(str(data['link_id'][i])[0:9])
    return data
```

### 27.오산시_도로명주소_건물.geojson

- building_cnt_27: 격자당 건물 개수
- building_level_up_27: 격자당 건물 지상층 수의 합
- building_level_down_27: 격자당 건물 지하층 수의 합
- bad_bld_27: 격자당 주변에서 교통사고가 많이 일어나는 건물 수
- good_bld_27: 격자당 주변에서 교통사고가 적게 일어나는 건물 수

```python
#전처리
df_27_1 = df_27[['BDTYP_CD','GRO_FLO_CO','UND_FLO_CO','geometry']]
df_27_1['geometry'] = df_27_1['geometry'].apply(multipolygon_to_center_coordinates)
df_27_1 = df_27_1[(df_27_1['BDTYP_CD']== '03102') | (df_27_1['BDTYP_CD']=='03103') | (df_27_1['BDTYP_CD']=='03104') | (df_27_1['BDTYP_CD']=='16002') | (df_27_1['BDTYP_CD']=='12002')].reset_index()
del df_27_1['index']
for i in range(len(df_27_1)):
    df_27_1['geometry'][i] = Point(df_27_1['geometry'][i][0][0], df_27_1['geometry'][i][0][1])
df_27_1.head(2)
```

```python
bad_bld = df_27[(df_27['BDTYP_CD']=='04016')|(df_27['BDTYP_CD']=='12000')|(df_27['BDTYP_CD']=='12001')|(df_27['BDTYP_CD']=='12002')|(df_27['BDTYP_CD']=='12008')|(df_27['BDTYP_CD']=='12009')|(df_27['BDTYP_CD']=='12999')|(df_27['BDTYP_CD']=='12000')|(df_27['BDTYP_CD']=='16001')]
good_bld = df_27[(df_27['BDTYP_CD']=='03102')|(df_27['BDTYP_CD']=='03103')]
```

```python
bad_bld['geometry'] = bad_bld['geometry'].apply(multipolygon_to_center_coordinates)
good_bld['geometry'] = good_bld['geometry'].apply(multipolygon_to_center_coordinates)
bad_bld = bad_bld.reset_index()
good_bld = good_bld.reset_index()

#multipolygon -> point
for i in range(len(bad_bld)):
    x = bad_bld['geometry'][i][0][0]
    y = bad_bld['geometry'][i][0][1]
    bad_bld['geometry2'] = Point(x,y)

for i in range(len(good_bld)):
    x = good_bld['geometry'][i][0][0]
    y = good_bld['geometry'][i][0][1]
    good_bld['geometry2'] = Point(x,y)

del bad_bld['geometry']
del good_bld['geometry']
bad_bld = bad_bld.rename(columns = {'geometry2':'geometry'})
good_bld = good_bld.rename(columns = {'geometry2':'geometry'})

bad_bld = bad_bld[['BDTYP_CD','geometry']]
good_bld = good_bld[['BDTYP_CD','geometry']]

bad_bld.head(2)
```

```python
#merge
#df_27 데이터는 point가 격자안에 있는지의 유무로 merge
#건물 개수
all_data['building_cnt_27'] = 0
for j in range(all_data['geometry'].shape[0]): #geometry
    bdcnt = 0
    for i in range(df_27_1['geometry'].shape[0]): #point
        pt = df_27_1['geometry'][i]
        if pt.within(all_data['geometry'][j]):
            bdcnt = bdcnt + 1
    all_data['building_cnt_27'][j] = bdcnt
```

```python
# df_27 데이터는 point가 격자안에 있는지의 유무로 merge
#건물지상층수, 건물 지하층수
all_data['building_level_up_27'] = 0
all_data['building_level_down_27'] = 0

for j in range(all_data['geometry'].shape[0]): #geometry
    bdup = 0
    bddn = 0
    for i in range(df_27_1['geometry'].shape[0]): #point
        pt = df_27_1['geometry'][i]
        if pt.within(all_data['geometry'][j]):
            bdup = bdup + df_27_1['GRO_FLO_CO'][i]
            bddn = bddn + df_27_1['UND_FLO_CO'][i]

    all_data['building_level_up_27'][j] = bdup
    all_data['building_level_down_27'][j] = bddn
```

```python
# df_27 데이터는 point가 격자안에 있는지의 유무로 merge
#교통사고 많이 나는 건물 수
all_data['bad_bld_27'] = 0

for j in range(all_data['geometry'].shape[0]): #geometry
    bcnt = 0
    for i in range(bad_bld['geometry'].shape[0]): #point
        pt = bad_bld['geometry'][i]
        if pt.within(all_data['geometry'][j]):
            bcnt = bcnt + 1

    all_data['bad_bld_27'][j] = bcnt
```
