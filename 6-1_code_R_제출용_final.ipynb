## 과제1. 어린이 보호구역 외 어린이 교통사고 위험지역 20개소 제시
### 6-1) 격자별 교통사고 위험지수 생성
> - **목표:** R 프로그래밍을 사용하여 최적 회귀 모형을 선정한 뒤 회귀계수를 이용하여 **격자별 교통사고 위험지수(hazard_index)**를 설정한다. 
> - **y: 어린이 교통 사고 횟수, X: 교통 사고 요인** 
> - **변수 선택:** step AIC로 주요 변수를 선택한 포아송 모형과 음이항 모형의 AIC를 비교하여 모형을 선택하고 그 모형에서 유의수준 0.05 이하의 유의한 변수를 선택한다.
> - **모형 선택:** **y가 사고 횟수로 영과잉 자료**이므로 영과잉 포아송 모형과 영과잉 음이항 모형을 **AIC로 비교하여 최적 모형을 선택**한다.
> - **위험 지수 설정:** **최적 모형의 회귀계수를 가중치로 사용**하여 격자별 교통사고 위험지수를 설정한다. 

```r
library(readxl)
library(MASS)
library(pscl)
library(xlsx)
```

```r
all_data = read_excel('input/all_data.xlsx')
all_data<-as.data.frame(all_data)
data<-all_data[,-c(1:6)]
data<- subset(data, select=-c(building_area_28))
head(data,3)
```

### 6-1-1) 포아송 모형

```r
poisson<-glm(accident_cnt_02~., data=data, family=poisson(link="log"))
summary(poisson)
```

```r
stepAIC(poisson) 
```

- step AIC를 통해 변수 선택 한 최종 포아송 모형 

```r
pois<-  glm(formula = accident_cnt_02 ~ parking_01 + pop_total_04 + pop_child_05 + 
    ratio_child_0405 + pop_working_06 + sign_cnt_16 + crosswalk_cnt_17 + 
    trafficlight_cnt_19 + cctv_cnt_20 + bus_cnt_22 + road_cnt_23 + 
    road_maxspeed_23 + road_length_23 + road_numcross_23 + building_level_down_27 + 
    sports_cnt_29 + bicycleroad_21 + pop_fluid_810_08 + school_cnt_10 + 
    sidewalk_cnt_21 + conjestion_freq_25 + intersection_23 + 
    barrier_23, family = poisson(link = "log"), data = data)

summary(pois)
```

```r
AIC(pois)
```

###  6-1-2) 음이항 모형

```r
nb<-glm.nb(accident_cnt_02~., data=data) 
summary(nb)
```

```r
stepAIC(nb) 
```

- step AIC를 통해 변수 선택 한 최종 음이항 모형 

```r
nbinom<-glm.nb(formula = accident_cnt_02 ~ parking_01 + pop_total_04 + 
    pop_child_05 + ratio_child_0405 + pop_working_06 + sign_cnt_16 + 
    crosswalk_cnt_17 + trafficlight_cnt_19 + cctv_cnt_20 + bus_cnt_22 + 
    road_maxspeed_23 + road_numcross_23 + building_level_down_27 + 
    sports_cnt_29 + bicycleroad_21 + pop_fluid_810_08 + school_cnt_10 + 
    sidewalk_cnt_21 + conjestion_freq_25 + intersection_23 + 
    barrier_23, data = data, init.theta = 1.031629867, link = log)
summary(nbinom)
```

```r
AIC(nbinom) 
```

## 분석 결과

#### step AIC로 주요 변수를 선택한 포아송 모형의 AIC는 2441, 음이항 모형의 AIC는 2387이므로 **AIC 기준으로 음이항 모형을 선택**하고 **유의수준  0.05 이하의 유의한 변수를 선택**한다.

#### 어린이 교통 사고 횟수를 설명하는데 유의한 변수로 **선택된 변수는 총 18개**이다.

- trafficlight_cnt_19: 격자내 신호등 개수 
- cctv_cnt_20: 격자내 cctv개수
- bus_cnt_22:격자내 버스정류장 개수 
- road_maxspeed_23: 격자내 도로의 제한속도 
- building_level_down_27: 격자내 건물의 지하층수
- sports_cnt_29: 격자내 체육시설 개수 
- pop_fluid_810_08: 2019년 오전 8~10시 격자내 유동인구

- conjestion_freq_25: 격자내 혼잡빈도강도  
- intersection_23: 격자내 교차로 개수 
- barrier_23: 격자내 중앙분리대 개수 

## 영과잉(Zero-inflated) 모델
- ZIP(Zero Inflated Poisson Regression), ZINB(Zero Inflated Negative Binomial Regression)
- 기존의 포아송회귀모형이나 음이항회귀모형과 같은 전통적인 추론방법은 특정한 값에서 기대도수보다 관측도수가 많이 나타나는 자료에 대하여 잘 설명할 수 없음
- 교통사고 자료와 같이 영의 비율이 기존의 포아송회귀모형 또는 음이항회귀모형에 의해 기대되는 영의 비율보다 높게 관측되는 자료를 영과잉 자료라고함.
- 영과잉 모형은 자료에서 과도하게 나타나는 영을 설명하기 위해 영에 대한 점확률을 가지는 분포를 기존의 분포와 합성한 것임

### 6-1-3) 영과잉 포아송 모형 (ZIP 모형)

```r
zip<-zeroinfl( accident_cnt_02 ~ parking_01 + pop_total_04 + 
    pop_child_05 + ratio_child_0405 + pop_working_06 + sign_cnt_16 + 
    crosswalk_cnt_17 + trafficlight_cnt_19 + cctv_cnt_20 + bus_cnt_22 + 
    road_maxspeed_23  + building_level_down_27 + 
    sports_cnt_29  + pop_fluid_810_08 + school_cnt_10 + conjestion_freq_25 + intersection_23 + 
    barrier_23 |1, data=data, dist="poisson")
summary(zip)
```

```r
AIC(zip)
```

### 6-1-4) 영과잉 음이항 모형 (ZINB 모형)

```r
zinb<-zeroinfl( accident_cnt_02 ~ parking_01 + pop_total_04 + 
    pop_child_05 + ratio_child_0405 + pop_working_06 + sign_cnt_16 + 
    crosswalk_cnt_17 + trafficlight_cnt_19 + cctv_cnt_20 + bus_cnt_22 + 
    road_maxspeed_23  + building_level_down_27 + 
    sports_cnt_29  + pop_fluid_810_08 + school_cnt_10 + conjestion_freq_25 + intersection_23 + 
    barrier_23 |1, data=data, dist="negbin")
summary(zinb)
```

```r
AIC(zinb)
```

### 분석 결과 
- step AIC로 주요 변수를 선택한 영과잉 포아송 모형의 AIC는 2437, 영과잉 음이항 모형의 AIC는 2394이므로 **AIC 기준으로 영과잉 음이항 모형을 최적 모형으로 선택**한다. 

### 6-1-5) 격자별 교통사고 위험지수 생성
> - 영과잉 음이항 모형을 통해 얻은 회귀 계수를 가중치로 사용하여 격자별 위험 지수 설정한다. 

```r
data_index<-all_data  
```

```r
# exponential 변환
data_index$hazard_index=exp(summary(zinb)$coefficients$count[1])+
                 exp(summary(zinb)$coefficients$count[3])*all_data$pop_total_04+
                 exp(summary(zinb)$coefficients$count[4])*all_data$pop_child_05+
                 exp(summary(zinb)$coefficients$count[5])*all_data$ratio_child_0405+
                 exp(summary(zinb)$coefficients$count[6])*all_data$pop_working_06 +
                 exp(summary(zinb)$coefficients$count[7])*all_data$sign_cnt_16+
                 exp(summary(zinb)$coefficients$count[8])*all_data$crosswalk_cnt_17+
                 exp(summary(zinb)$coefficients$count[9])*all_data$trafficlight_cnt_19+
                 exp(summary(zinb)$coefficients$count[10])*all_data$cctv_cnt_20+
                 exp(summary(zinb)$coefficients$count[11])*all_data$bus_cnt_22 +
                 exp(summary(zinb)$coefficients$count[12])*all_data$road_maxspeed_23+
                 exp(summary(zinb)$coefficients$count[13])*all_data$building_level_down_27+
                 exp(summary(zinb)$coefficients$count[14])*all_data$sports_cnt_29 +
                 exp(summary(zinb)$coefficients$count[15])*all_data$pop_fluid_810_08 +
                 exp(summary(zinb)$coefficients$count[16])*all_data$school_cnt_10+
                 exp(summary(zinb)$coefficients$count[17])*all_data$conjestion_freq_25+
                 exp(summary(zinb)$coefficients$count[18])*all_data$intersection_23+
                 exp(summary(zinb)$coefficients$count[19])*all_data$barrier_23 
```

```r
write.xlsx(data_index,'input/data_index.xlsx')  # 위험지수를 포함한 데이터를 저장
```
